```python
# for filling the missing months for the plots 

def missing_months(number_of_missing_months, data):
    import array
    a = np.array(data)
    aa = []
    
    for i in a:
        aa.append(round(i))
        
    b = np.zeros(number_of_missing_months)
    
    bb = []
    for i in b:
        bb.append(int(i))
    
    arr1 = array.array('i', aa)
    arr2 = array.array('i', bb)
    week_2023_arr3 = arr1 + arr2
    
    return week_2023_arr3


from scipy.stats import norm

def norm_fun(x):
    x = x
    std = np.std(x)
    mean = np.mean(x)
    
    y = norm.pdf(x,mean,std)
    
    plt.plot(x,y,'o')
    plt.hist(x, bins=30, density=True, alpha=0.6, color='b') 
    plt.show()

```

```python
def cumplot(data, groupby_): # groupby_ = Day, Minutes, week 
    
    df = data

    # Define variables to plot
    y_mean = df.groupby(str(groupby_)).mean()['spikes_height']
    x = y_mean.index


    # Compute upper and lower bounds using chosen uncertainty measure: here
    # it is a fraction of the standard deviation of measurements at each
    # time point based on the unbiased sample variance
    y_std =df.groupby(str(groupby_)).std()['spikes_height']
    error = y_std
    lower = y_mean - error
    upper = y_mean+ error
    
    return x,y_mean,lower,upper
    
```

```python
def cumplot_cummulative(data, groupby_): # groupby_ = Day, Minutes, week 
    
    df = data

    # Define variables to plot
    y_mean = df.groupby(str(groupby_)).mean()['C_anc_gust_wind_speed']
    x = y_mean.index

    # Compute upper and lower bounds using chosen uncertainty measure: here
    # it is a fraction of the standard deviation of measurements at each
    # time point based on the unbiased sample variance
    y_std =df.groupby(str(groupby_)).std()['C_anc_gust_wind_speed']
    
    error = 0.5*y_std

    lower = data['C_anc_gust_wind_speed'].min()
    upper =data['C_anc_gust_wind_speed'].max()
    
    
    return x,y_mean,lower,y_std
```
