```python
def select_date_weather(start_day, end_day,start_time,end_time,data_kind): # data_kind = rainfall, temperature, or pressure
    
        # Unlock here to mask the data
        
    if data_kind== 'temperature':

        df_2020 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_temperature_2020.pkl')
        df_2021= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_temperature_2021.pkl')
        df_2022 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_temperature_2022.pkl')
        df_2023= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_temperature_2023.pkl')

    elif data_kind == 'rainfall':

        df_2020 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_weather_rain_2020.pkl')
        df_2021= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_weather_rain_2021.pkl')
        df_2022 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_weather_rain_2022.pkl')
        df_2023= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_weather_rain_2023.pkl')
        
    elif data_kind == 'pressure':

        df_2020 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_pressure_2020.pkl')
        df_2021= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_pressure_2021.pkl')
        df_2022 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_pressure_2022.pkl')
        df_2023= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_pressure_2023.pkl')
        
    elif data_kind == 'humidity':

        df_2020 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_relative_humidity_2020.pkl')
        df_2021= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_relative_humidity_2021.pkl')
        df_2022 = pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_relative_humidity_2022.pkl')
        df_2023= pd.read_pickle('/data/Wind_data_project/Meerkat_data/anc_air_relative_humidity_2023.pkl')


    start = start_day
    end = end_day

    start_date = '2020-'+start+'-'+ start_time
    end_date = '2020-'+end+'-'+end_time 
    

    mask_2020 = (df_2020['Timestamp'] > start_date) & (df_2020['Timestamp'] <= end_date)
    df1_2020_mask = df_2020.loc[mask_2020]

    start_date = '2021-'+start+'-'+ start_time
    end_date = '2021-'+end+'-'+end_time 

    mask_2021 = (df_2021['Timestamp'] > start_date) & (df_2021['Timestamp'] <= end_date)
    df1_2021_mask = df_2021.loc[mask_2021]

    start_date = '2022-'+start+'-'+ start_time
    end_date = '2022-'+end+'-'+end_time 

    mask_2022 = (df_2022['Timestamp'] > start_date) & (df_2022['Timestamp'] <= end_date)
    df1_2022_mask = df_2022.loc[mask_2022]

    start_date = '2023-'+start+'-'+ start_time
    end_date = '2023-'+end+'-'+end_time 

    mask_2023 = (df_2023['Timestamp'] > start_date) & (df_2023['Timestamp'] <= end_date)
    df1_2023_mask = df_2023.loc[mask_2023]

    df_2020 = df1_2020_mask
    df_2021 = df1_2021_mask
    df_2022 = df1_2022_mask
    df_2023 = df1_2023_mask
    
    
    
    
    df1_2020 = df_2020
    df1_2020["New_Date"] = pd.to_datetime(df_2020["Timestamp"]).dt.date
    df1_2020["New_Time"] = pd.to_datetime(df_2020["Timestamp"]).dt.time
    df1_2020["Month"] = pd.to_datetime(df_2020["Timestamp"]).dt.month
    df1_2020["week"] = pd.to_datetime(df_2020["Timestamp"]).dt.week
    df1_2020["Minutes"] = pd.to_datetime(df_2020["Timestamp"]).dt.minute
    df1_2020["Day"] = pd.to_datetime(df_2020["Timestamp"]).dt.day
    df1_2020["Secounds"] = pd.to_datetime(df_2020["Timestamp"]).dt.second
    df1_2020["Hour"] = pd.to_datetime(df_2020["Timestamp"]).dt.hour
    df1_2020 = df1_2020.set_index('Timestamp')
    df1_2020.index = pd.to_datetime(df1_2020.index)



    df1_2021 = df_2021
    df1_2021["New_Date"] = pd.to_datetime(df_2021["Timestamp"]).dt.date
    df1_2021["New_Time"] = pd.to_datetime(df_2021["Timestamp"]).dt.time
    df1_2021["Month"] = pd.to_datetime(df_2021["Timestamp"]).dt.month
    df1_2021["week"] = pd.to_datetime(df_2021["Timestamp"]).dt.week
    df1_2021["Minutes"] = pd.to_datetime(df_2021["Timestamp"]).dt.minute
    df1_2021["Day"] = pd.to_datetime(df_2021["Timestamp"]).dt.day
    df1_2021["Secounds"] = pd.to_datetime(df_2021["Timestamp"]).dt.second
    df1_2021["Hour"] = pd.to_datetime(df_2021["Timestamp"]).dt.hour
    df1_2021 = df1_2021.set_index('Timestamp')
    df1_2021.index = pd.to_datetime(df1_2021.index)
    
    
    df1_2022 = df_2022
    df1_2022["New_Date"] = pd.to_datetime(df_2022["Timestamp"]).dt.date
    df1_2022["New_Time"] = pd.to_datetime(df_2022["Timestamp"]).dt.time
    df1_2022["Month"] = pd.to_datetime(df_2022["Timestamp"]).dt.month
    df1_2022["week"] = pd.to_datetime(df_2022["Timestamp"]).dt.week
    df1_2022["Minutes"] = pd.to_datetime(df_2022["Timestamp"]).dt.minute
    df1_2022["Secounds"] = pd.to_datetime(df_2022["Timestamp"]).dt.second
    df1_2022["Hour"] = pd.to_datetime(df_2022["Timestamp"]).dt.hour
    df1_2022["Day"] = pd.to_datetime(df_2022["Timestamp"]).dt.day
    df1_2022 = df1_2022.set_index('Timestamp')
    df1_2022.index = pd.to_datetime(df1_2022.index)
    
    
    df1_2023 = df_2023
    df1_2023["New_Date"] = pd.to_datetime(df_2023["Timestamp"]).dt.date
    df1_2023["New_Time"] = pd.to_datetime(df_2023["Timestamp"]).dt.time
    df1_2023["Month"] = pd.to_datetime(df_2023["Timestamp"]).dt.month
    df1_2023["week"] = pd.to_datetime(df_2023["Timestamp"]).dt.week
    df1_2023["Minutes"] = pd.to_datetime(df_2023["Timestamp"]).dt.minute
    df1_2023["Day"] = pd.to_datetime(df_2023["Timestamp"]).dt.day
    df1_2023["Secounds"] = pd.to_datetime(df_2023["Timestamp"]).dt.second
    df1_2023["Hour"] = pd.to_datetime(df_2023["Timestamp"]).dt.hour
    df1_2023 = df1_2023.set_index('Timestamp')
    df1_2023.index = pd.to_datetime(df1_2023.index)
    
    
    
    
    
    return df1_2020,df1_2021,df1_2022,df1_2023
```

```python
def cumplot(data, groupby_,data_kind): # groupby_ = Day, Minutes, week 
    
    df = data
    if data_kind == 'temperature':

        # Define variables to plot
        y_mean = df.groupby(str(groupby_)).mean()['anc_air_temperature']
        x = y_mean.index

        y_std =df.groupby(str(groupby_)).std()['anc_air_temperature']
        error = 2*y_std
        lower = df.groupby(str(groupby_)).min()['anc_air_temperature']
        upper = df.groupby(str(groupby_)).max()['anc_air_temperature']

    
    elif data_kind== 'rainfall':
        
        # Define variables to plot
        y_mean = df.groupby(str(groupby_)).sum()['anc_weather_rain']
        x = y_mean.index

        y_std =df.groupby(str(groupby_)).std()['anc_weather_rain']
        error = 2*y_std
        lower = df.groupby(str(groupby_)).min()['anc_weather_rain']
        upper = df.groupby(str(groupby_)).max()['anc_weather_rain']
        
    
    elif data_kind== 'pressure':
        
        # Define variables to plot
        y_mean = df.groupby(str(groupby_)).mean()['anc_air_pressure']
        x = y_mean.index

        y_std =df.groupby(str(groupby_)).std()['anc_air_pressure']
        error = 2*y_std
        lower = df.groupby(str(groupby_)).min()['anc_air_pressure']
        upper = df.groupby(str(groupby_)).max()['anc_air_pressure']
        
    elif data_kind== 'humidity':
        
        # Define variables to plot
        y_mean = df.groupby(str(groupby_)).mean()['anc_air_relative_humidity']
        x = y_mean.index
        
        y_std =df.groupby(str(groupby_)).std()['anc_air_relative_humidity']
        error = 2*y_std
        lower = df.groupby(str(groupby_)).min()['anc_air_relative_humidity']
        upper = df.groupby(str(groupby_)).max()['anc_air_relative_humidity']
        
        


    return x,y_mean, y_mean-y_std ,y_mean+y_std,y_std
```

```python



data_kind = ['rainfall','pressure','temperature','humidity']

start = start_day
end = end_day



for kind in data_kind: 
    
    
    df1_2020,df1_2021,df1_2022,df1_2023= select_date_weather(start_day, end_day,start_time,end_time,kind)  # data_kind = rainfall or temperature

    group_by_ = 'week'

    group_2020 = cumplot(df1_2020,group_by_,kind)
    group_2021 = cumplot(df1_2021,group_by_,kind)
    group_2022 = cumplot(df1_2022,group_by_,kind)
    group_2023 = cumplot(df1_2023,group_by_,kind)
    
    fig, ax = plt.subplots(4,figsize=(20, 15),sharex=True)
    fig.tight_layout(pad=5.0)
    
    ax[0].errorbar(group_2020[0],group_2020[1],group_2020[4],clip_box ='Bbox', color='b',label='2020')
    ax[1].errorbar(group_2021[0],group_2021[1],group_2021[4], color='r',label='2021')
    ax[2].errorbar(group_2022[0],group_2022[1],group_2022[4], color='g',label='2022')
    ax[3].errorbar(group_2023[0],group_2023[1],group_2023[4], color='m',label='2023')

    ax[0].fill_between(group_2020[0],group_2020[2], 
                 group_2020[3], color='b',alpha=0.25)
    ax[1].fill_between(group_2021[0],group_2021[2], 
                 group_2021[3], color='r',alpha=0.25)
    ax[2].fill_between(group_2022[0],group_2022[2], 
                 group_2022[3], color='g',alpha=0.25)
    ax[3].fill_between(group_2023[0],group_2023[2], 
                 group_2023[3], color='m',alpha=0.25)



    if kind== 'rainfall':
        
        fig.suptitle('Total rainfall per week',fontsize=20)
        
        #ax[0].set_title('Total rainfall per week',fontsize=20,fontweight='bold')
        ax[0].set_ylabel('Rainfall (mm)',fontsize=16,fontweight='bold')
        ax[1].set_ylabel('Rainfall (mm)',fontsize=16,fontweight='bold')
        ax[2].set_ylabel('Rainfall (mm)',fontsize=16,fontweight='bold')
        ax[3].set_ylabel('Rainfall (mm)',fontsize=16,fontweight='bold')
        
        ax[3].set_xlabel(group_by_,fontsize=16)

        ax[0].legend(fontsize=16)
        ax[1].legend(fontsize=16)
        ax[2].legend(fontsize=16)
        ax[3].legend(fontsize=16)
        
        ax[0].set_xticks(range(0, 54,2))
        ax[1].set_xticks(range(0, 54,2))
        ax[2].set_xticks(range(0, 54,2))
        ax[3].set_xticks(range(0, 54,2))
        
        plt.savefig('Weekly_weather_rain_sep')


    elif kind== 'pressure':
        
        fig.suptitle('Average air pressure per week',fontsize=20)
            
        
        #ax[1].set_title('Average air pressure per week',fontsize=20,fontweight='bold')
        ax[0].set_ylabel('Air pressure',fontsize=16,fontweight='bold')
        ax[1].set_ylabel('Air pressure',fontsize=16,fontweight='bold')
        ax[2].set_ylabel('Air pressure',fontsize=16,fontweight='bold')
        ax[3].set_ylabel('Air pressure',fontsize=16,fontweight='bold')
        
        ax[3].set_xlabel(group_by_,fontsize=16,fontweight='bold')
        
        ax[0].legend(fontsize=16)
        ax[1].legend(fontsize=16)
        ax[2].legend(fontsize=16)
        ax[3].legend(fontsize=16)
        
        ax[0].set_xticks(range(0, 54,2))
        ax[1].set_xticks(range(0, 54,2))
        ax[2].set_xticks(range(0, 54,2))
        ax[3].set_xticks(range(0, 54,2))
        
        plt.savefig('Weekly_weather_pressure_sep')
        
        
    elif kind== 'humidity':
        
        fig.suptitle('Average humidity per week ',fontsize=20)
        

        #ax[2].set_title('Average humidity per week ',fontsize=20,fontweight='bold')
        ax[0].set_ylabel('Air relative humidity',fontsize=16,fontweight='bold')
        ax[1].set_ylabel('Air relative humidity',fontsize=16,fontweight='bold')
        ax[2].set_ylabel('Air relative humidity',fontsize=16,fontweight='bold')
        ax[3].set_ylabel('Air relative humidity',fontsize=16,fontweight='bold')
        
        
        
        ax[3].set_xlabel(group_by_,fontsize=16,fontweight='bold')
        
        ax[0].legend(fontsize=16)
        ax[1].legend(fontsize=16)
        ax[2].legend(fontsize=16)
        ax[3].legend(fontsize=16)
        
        ax[0].set_xticks(range(0, 54,2))
        ax[1].set_xticks(range(0, 54,2))
        ax[2].set_xticks(range(0, 54,2))
        ax[3].set_xticks(range(0, 54,2))
        plt.savefig('Weekly_weather_humidity_sep')




# #plt.savefig('Weekly_weather')
plt.show()
```
